```python
import pandas as pd
```

```python
yelp_raw_data = pd.read_csv('yelp.csv')
```

Let's use a method called value_counts to see the count
distribution

```python
dates = yelp_raw_data['stars'].value_counts()
dates
```

We can also plot this data to get a nice visual. First,
let's sort by star rating, and then use the prebuilt plot method to make a bar chart

```python
sorted(dates)
```

```python
dates.plot(kind = 'bar')
```
