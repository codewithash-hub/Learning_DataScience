```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
bikes = pd.read_csv('bikeshare.csv')
bikes.head()
```

```python
bikes.plot(kind = 'scatter', x ='temp', y = 'count', alpha = 0.2)
```

```python
sns.lmplot(x = 'temp', y = 'count', data = bikes, aspect = 1.5, scatter_kws = {'alpha': .2})
```

```python
bikes[['count', 'temp']].corr()
```

```python
import numpy as np
```

```python
feature_cols = ['temp']
x = np.array(bikes[feature_cols]).reshape(-1, 1)
y = bikes['count']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(x, y)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python

linreg.predict(20)
```
