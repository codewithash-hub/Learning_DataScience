# Predict the number of bikes on a particular day for a bike sharing program

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
bikes = pd.read_csv('bikeshare.csv')
bikes.head()
```

```python
# bikes.plot(kind = 'scatter', x = 'temp', y = 'count', alpha = 0.2)
sns.lmplot(x = 'temp', y = 'count', data = bikes, aspect = 1.8, scatter_kws = {'alpha': 0.2})

# x = bikes['temp']
# y = bikes['count']

# plt.scatter(x, y, alpha = 0.2)
```

```python
bikes[['count', 'temp']].corr()
```

```python
from sklearn.linear_model import LinearRegression

feature_cols = ['temp']
x = bikes[feature_cols]
y = bikes['count']

model = LinearRegression()
model.fit(x, y)

print(model.intercept_)
print(model.coef_[0])

# print(model.predict(20))
```
